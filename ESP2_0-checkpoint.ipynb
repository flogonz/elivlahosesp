```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def slopes (t):
    return (c*r*t)/k - g/(2*r*(r-k))*((m-r*t)**2)-(((m-r*t)**(1+k/r))/(r*(1+k/r)))*((-c*r)*(r-k)-g*m*k)/(k*(r-k)*m**(k/r))+g*m**2/(2*r*(r-k))+m**(1+k/r)/(r*(1+k/r))*((-c*r)*(r-k)-g*m*k)/(k*(r-k)*m**(k/r))
```

```python
def slope(t, y):
    return [y[1], -(k*y[1])/(m-r*t) + (c*r)/(m-r*t) - (e*G)/((d)**2)]
```

```python
y0 = np.array([0, 0])
t0 = 0
tf = 120
G = 6.67*10**(-11)
g = 9.8
k = 1.13
m = 231870
d = 6378100
e = 5.972*10**(24)
r = 1816
c = 5000 ##2450?
ts = (t0, tf)
n = 1000
T = np.linspace(t0, tf, n)
```

```python
sol = solve_ivp(slope, ts, y0, method="RK45", t_eval=T)
```

```python
plt.xlabel("t")
plt.ylabel("y")
plt.title("Position of a rocket over time")
plt.plot(sol.t, sol.y[0], "r") 
plt.show()
```

```python
plt.xlabel("t")
plt.ylabel("y")
plt.title("Position of a rocket over time")
plt.plot(sol.t, slopes(T), "r") 
plt.show()
```

```python
plt.plot(sol.t, slopes(T)-sol.y[0], 'b')
plt.xlabel('t')
plt.ylabel('Y-y')
plt.suptitle("Difference between analytical and numerical solver")
plt.show()
```
